# Introduction to Supervised Learning

Supervised machine learning means **predicting $`y`$ from $`\vec{x}`$** from a model trained on a dataset. It is supervised because we **tell the algorithm what the labels are** in the data set, and the models predictions are **supervised by an error function/critic**.

As an introduction to supervised learning, we will use a dataset called **AqSolDB**. **<sup>1</sup>** The AqSolDB datset contains ~10,000 unique compounds with the **label for each data point/compound** being their **solubility in water**. 
* The chosen unit for solubility in this dataset is **LogS**, where **S** is the the aqueous solubility in **mol/dm<sup>3</sup> (mol/L)**

The dataset also contains **molecular properties (features/attributes)** that we can use for machine learning.

# 4) Linear Regression

We will begin with one of the **simplest supervised learning algorithms** - **linear regression**.

The linear regression algorithm is defined by the equation:

$$y = \vec{w} \cdot \vec{x} + b$$

where:
- $Y$ is the **predicted output (dependent variable)**
- $\vec{x}$ is the **a input feature vector (independent variables)**. The shape of a single feature vector in our case is **17** as there are **17 features**.
- $\vec{w}$ is the **weight vector (adjustable parameters**. In our case it is of **length 17**.
- $b$ is the **bias term (the intercept)**. This is an **adjustable scalar**.

In this model, the goal is to **predict $Y$** based on the **input features $\vec{x}$** by **adjusting the weights $\vec{w}$ and bias $b$**.

We will implement the linear regression model using a library called **`jax`**. This library is very similar to  `numpy`, the main difference is that is can **compute analytical gradients** easily via **autodiff**.
* Autodiff is a tool that can **compute analytical gradients** with respect to **two variables**

````python
# Test linear regression utility function
linear_regression_utils.linear_regression(
    feature_vector=np.array([1, 0, 2.5]),
    weight_vector=np.array([0.2, -0.5, 0.4]),
    bias_term = 4.3
)
````

## 4.1) Calculating $\vec{w}$ and $b$

We know what $\vec{x}$ will be as it will be a **numpy array of all the features in the dataset**. The critical question however is, **how do we determine the adjustable parameters $\vec{w}$ and $b$?**

The classic solution for calculating the adjustable parameters in linear regression is to use the **least-squares method**.
* The least-squares method involves finding the parameters $\vec{w}$ that **minimise the sum of the squared differences between the observed and predicted values**.

This can be achieved with **matrix algebra** via a **psuedo-inverse**:

$$\vec{w} = (X^TX)^{-1}X^{T}\vec{y}$$

where:
- $X$ is the **design matrix**, where **each row corresponds to an input feature vector $\vec{x}$**. If there are **$n$** samples and **$m$** features, **$X$** is an **$n \times m$ matrix**.
- $\vec{y}$ is the **vector of the observed output values/predictions**
- $X^T$ is the **transpose of the design matrix $X$**
- $(X^T X)^{-1}$ is the **inverse product of $X^T$ and $X$. This term is known as the **psuedo-inverse of $X$**.

In our example we will use an **iterative approach** that mirrors what is done in deep learning. This is **not the correct approach for linear regression**, but it will be useful to cover the iterative approach as it **used so often in deep learning**.

To **iteratively find our adjustable parameters**, we will pick a **loss function** and **minimise the gradients**.
* A **loss** is a function which **takes in a model prediction **$\vec{y}$**, labels **$y$** and **calculates a scalar representing how poor the fit is**.
* Our goal is to **minimise loss**

````python
# Extract features and labels from dataset
aq_sol_feature_values = aq_sol_dataset.loc[:, aq_sol_dataset_feature_names].values
aq_sol_label_values = aq_sol_dataset.Solubility.values

# Calculate number of features (feature dimensionality)
aq_sol_feature_dim = aq_sol_feature_values.shape[1]

# Initialise a the weight vector with random values from a normal distribution. 
# The size of the vector is equal to the number of features
weight_vector = np.random.normal(size=aq_sol_feature_dim)

# Initialise the bias term as 0.0
bias_term: float = 0.0

linear_regression_utils.mean_squared_error_loss_wrapper(
    weight_vector,
    bias_term,
    aq_sol_feature_values,
    aq_sol_label_values
)
````

## 4.2) Gradient Descent

The value returned by the mean squared error loss function is **very large**. We will now attempt to **reduce loss** by **using information about how it changes with respect to the adjustable parameters**.

### 4.2.1) Mean Squared Loss Function
Our mean squared loss function **measures the average squared difference between the predicted and true values** and is defined as:

$$L = \frac{1}{N}\sum_i^N \left[y_i - f(\vec{x}_i, \vec{w}, b)\right]^2$$

where:
- $L$ is the **loss function**, specifically the mean squared error in this case.
- $N$ is the **number of data points**.
- $y_i$ is the **true output for the $i$-th data point**.
- $f(\vec{x}_i, \vec{w}, b)$ is the **predicted value for the $i$-th data point**, as predicted by the **linear regression model**.

### 4.2.2) Gradients of the Loss Function
To **minimise the loss**, we need to know **how the loss changes with respect to each parameter**. This is achieved by **calculating the gradients (partial derivatives) of the loss function**:

$$\frac{\partial L}{\partial w_i}, \frac{\partial L}{\partial b}$$

where:
- $\frac{\partial L}{\partial w_i}$ is the **gradient of the loss function with respect to the $i$-th weight $w_i$** of the weight vector $\vec{w}$. This tells us **how the loss changes as we adjust the $i$-th weight**.
- $\frac{\partial L}{\partial b}$ is the **gradient of the loss function with respect to the bias term $b$**. This tells us **how the loss changes as we adjust the bias term**

### 4.2.3) Gradient Descent Update Rule
Gradient descent is a method to **iteratively update the parameters to minimize the loss**. The **update rule** for the parameters is given by:

$$(w_i, b') = \left(w_i - \eta \frac{\partial L}{\partial w_i}, b - \eta\frac{\partial L}{\partial b}\right)$$

where:
- $\eta$ is the **learning rate**. This is a **hyperparameter** that **determines the step size for each update**. It **controls how much we adjust the parameters in the direction of the negative gradient**.

We update the parameters in the **direction of the negative gradient** as this is the **direction that reduces loss**

* The learning rate $\eta$ is **not learned from the data**. Instead it is **set before training begins**. It is usually a **small value**, often chosen as a power of 10 (e.g. $10^-6$).
* The size of the learning rate matters, **too low of a learning rate** can cause the training process to be **unstable/low quality**, whereas a **too high of a learning rate** can make the training process **very slow**.

### 4.2.4) Linear Regression & Loss Function Optimisation Overall Steps

Putting all these steps together:
1. **Initialisation**: Start with initial guesses for the weights $\vec{w}$ and bias $b$.
2. **Calculate Predictions**: Use linear regression to calculate predictions for all data points.
3. **Calculate Loss**: Calculate the MSE loss using the predictions and true values.
4. **Calculate Gradients**: Calculate the gradients of the loss with respect to each parameter (the weights and the bias).
5. **Update Parameters**: Adjust the parameters using the gradient descent update rule.
6. **Repeat**: Repeat steps 4 and 5 iteratively until the loss converges to a minimum or a predefined number of iterations is reached.

In order to calculate the partial derivative/gradient we will use `jax.grad`. This calculates an **analytical derivative** of a Python function. It takes **two arguments**: the **function** and **which args to take the derivative of**. 
* For example, for function `f(x, y, z)`, `jax.grad(f, (0, 1))` would give $\frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}$.

### 4.2.5) Calculate Gradient with Respect to Weight Vector ($\vec{w}$) and Bias Term ($b$)

````python
# Instantiate `loss_grad` which is an instance of `jax.grad()` that calculates 
# the analytical derivative of arguments 0 (weight vector) and 1 (bias term) of the 
# `mean_squared_error_loss_wrapper()` function

loss_grad = jax.grad(linear_regression_utils.mean_squared_error_loss_wrapper, (0, 1))

loss_grad(
    weight_vector, 
    bias_term, 
    aq_sol_feature_values, 
    aq_sol_label_values
)
````

### 4.2.6) Output of `jax.grad()`

The output of `loss_grad` is a **tuple containing two elements**:
1. An Array containing the **gradient of the loss with respect to the weight vector**. This is:
   ```
   Array([-5.21945562e+05, -3.51245288e+03, -1.30674188e+05, -3.46420625e+04,
        -7.20934473e+03, -1.83106812e+03, -1.03583438e+04, -7.93461426e+03,
        -1.83170297e+05, -2.49595996e+03, -4.81597107e+02, -8.47923950e+02,
        -3.34388013e+03, -1.21781836e+05, -2.16472078e+05, -2.45878931e+03,
        -1.19926288e+06], dtype=float32)
   ```
   * The **length of this array matches the number of features in the input data**
   * Each element in this array represents **how much the loss would change with a small change in the corresponding weight**
   * For example, the 1st element `-5.21945562e+05` indicates that **increasing the first weight by a small amount would decrease the loss by approximately `-5.21945562e+05` units**
2. An Array containing the **gradient of the loss with respect to the bias term**. This is:
   ```
   Array(-1213.9989, dtype=float32, weak_type=True)
   ```
   * This single element array indicates **how much the loss would change with a small change in the bias term**
   * For example, the value `-1213.9989` means that **increasing the bias terma by a small amount would decrease the loss by approximately `-1213.9989` units**

## 4.3) Training Curve (Learning Curve)

In machine learning, a **training curve** or **learning curve** plots the **value of a model's loss function** for a training set over a **number of optimisation/training steps**. It is a **measure of a model's predictive performance** over an iterative number of optimisation/training steps.

Generally, **if the loss functions value is decreasing** over the optimisation/training steps, then the **model is learning**.

Below we do a **100 step optimisation/training** of our linear regression model and plot the training/learning curve.

````python
loss_progress = []
learning_rate = 1e-6
for i in range(100):
    # Calculate gradients w.r.t weights and bias term
    grad = loss_grad(
        weight_vector, 
        bias_term, 
        aq_sol_feature_values, 
        aq_sol_label_values
    )

    # Update weight vector by subtracting the product of the learning rate and the gradient
    # w.r.t the weights. This moves the weights in the direction of the negative gradient,
    # reducing the loss
    weight_vector -= learning_rate  * grad[0]

    # Similarly, update bias term by subtracting the product of the learning rate and the 
    # gradient w.r.t the bias
    bias_term -= learning_rate  * grad[1]

    # Calculate new MSE loss on each iteration to track loss progress
    loss_progress.append(
        linear_regression_utils.mean_squared_error_loss_wrapper(
            weight_vector, 
            bias_term,
            aq_sol_feature_values,
            aq_sol_label_values
        )
    )
    
plt.plot(loss_progress)
plt.xlabel("Step")
plt.yscale("log")
plt.ylabel("Loss")
plt.title("AqSolDB Training Curve")
plt.show()
````

## 4.4) Batching - Stochastic Gradient Descent (SGD)

In reality, **training is not done on the full dataset all at once**. Instead, a **small batch of data** is used in a process called **batching**. 

This approach offers two benefits:
1. It **reduces the time needed to compute parameter updates**.
2. It **introduces randomness to the training process**. This randomness makes it possible to **escape local minima** which might otherwise **halt training progress**.

This introduction of randonmess with batching makes our algorithm **stochastic** and so this new gradient descent is called **stochastic gradient descent (SGD)**. 
* In the context of machine learning and optimisation, **stochastic** refers to **processes that involve some level of randomness or probabilistic behaviour**.
* When we say an algorithm is stochastic, we mean that it **incorporates randomness in its execution**, leading to **variability in its outcomes**. even when it is **run multiple times on the same dataset**.

SGD, and variations of it, are the **most common methods of training in deep learning**.

````python
# Initialise a the weight vector with random values from a normal distribution. 
# The size of the vector is equal to the number of features
weight_vector = np.random.normal(size=aq_sol_feature_dim)

# Initialise the bias term as 0.0
bias_term: float = 0.0

loss_progress: List = []
learning_rate = 1e-6
batch_size: int = 32
num_data_points: int = len(aq_sol_label_values)

# Prepare batches
#
# Calculate largest multiple of batch size that fits into dataset, ensuring data is evenly
# distributed into batches, and dropping any left over data. This leads to more consistent
# and stable updates
num_batch_data = num_data_points // batch_size * batch_size

# Reshape `aq_sol_feature_values` and `aq_sol_label_values` matrices/DataFrames to have 
# dimensions of `(-1, batch_size, feature_dim)`. The `-1` allows Numpy to automatically
# calculate the appropriate size for this dimension
aq_sol_batched_features = aq_sol_feature_values[:num_batch_data].reshape(
    (-1, batch_size, aq_sol_feature_dim)
)
aq_sol_batched_labels = aq_sol_label_values[:num_batch_data].reshape(
    (-1, batch_size)
)

# Iterate over batches randomly for stochastic gradient descent
indices = np.arange(num_batch_data // batch_size)
np.random.shuffle(indices)

for i in indices:
    # Calculate gradients w.r.t weights and bias term
    grad = loss_grad(
        weight_vector, 
        bias_term, 
        aq_sol_batched_features[i], 
        aq_sol_batched_labels[i]
    )

    # Update weight vector by subtracting the product of the learning rate and the gradient
    # w.r.t the weights. This moves the weights in the direction of the negative gradient,
    # reducing the loss
    weight_vector -= learning_rate  * grad[0]

    # Similarly, update bias term by subtracting the product of the learning rate and the 
    # gradient w.r.t the bias
    bias_term -= learning_rate  * grad[1]

    # Calculate new MSE loss on each iteration to track loss progress every 10 steps
    if i % 10 == 0:
        loss_progress.append(
            linear_regression_utils.mean_squared_error_loss_wrapper(
                weight_vector, 
                bias_term,
                aq_sol_feature_values,
                aq_sol_label_values
            )
        )

# Scale x-axis coordinates by 10 to include every step as we only record loss every 10 steps. 
# Keep y-axis coordinates the same
plt.plot(np.arange(len(loss_progress)) * 10, loss_progress)
plt.xlabel("Step")
plt.yscale("log")
plt.ylabel("Loss")
plt.title("Batched AqSolDB Training Curve")
plt.show()
````

### 4.4.1) Gradient Descent vs Stochastic Gradient Descent (Not Batching vs Batching)

We can see three changes from using batching:
1. The **reduction in loss is greater than without batching**:
   * This is because we **update the parameters more frequently**. With batching, we perform parameter updates **at each batch**, leading to **more parametet updates per iteration over the whole dataset**.
   * Specifically, if $B$ is the batch size and $N$ is the total number of data points, there are **$N/B$ updates for every 1 update in the original gradient descent**
2. There are **more steps**, even though we **iterated over our dataset once** instead of 100 times.
   * This is the same reason as the greater reduction in loss. **More updates == more steps**
3. The **loss doesn't always go down**.
   * The loss doesn't always go down because **each update/step is based on a different subset of the dataset**. Some data points/certain molecules are **harder to predict that others**.
   * Additionally, **each update is based on one batch**, which **might not represent the entire dataset well**. However, assuming our **batches are mixed well**, stochastic gradient descent will **always reduce loss on average**.

## 4.5 ) Standardise Features

We appear to have **hit a limit with our current loss**. If we **examine the gradients**, we notice that **some are very large** while **others are very small**.

This is because the **the values for different features have different magnitudes**. For example, **molecular weights are large numbers**, whereas the **number of rings in a molecule is a small number**.

Therefore, **using the same learning rate $\eta$ for all features works well for some** but it **too small for others**. If, however, we were to **increase $\eta$, the training process would become unstable due to these large feature gradients for molecular weight**, for example.

A standard approach to handle this issue is to **standardise the features** so that they have **similar magnitudes**. This can be done using the **standardisation equation commonly found in statistics**:

$$x_{ij} = \frac{x_{ij} - \bar{x_j}}{\sigma_{x_j}}$$

where:
- $x_ij$ is the **$j$-th feature for the $i$-th data point**.
  * $i$ is the **index of the data point** in the training dataset.
  * $j$ is the **index of the feature (column) in the training dataset**
- $\bar{x_j}$ is the **mean value of the $j$-th feature across all data points in the training dataset**. Also known as the **column average**:
$$\bar{x_j} = \frac{1}{N} \sum_{i=1}^{N} x_{ij}$$
- $\sigma_{x_j}$ is the **standard deviation of the $j$-th feature across all data points in the training dataset**. Also known as the **column standard deviation**:
$$\sigma_{x_j} = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (x_{ij} - \bar{x_j})^2}$$

To **avoid contaminating the training data with test data**, we must calculate the column mean and column standard devation **using only the training data**. This **prevents leaking information between the training data and test datasets**.

This is because our **test data should reflect how the model will perform on unseen data**. We therefore **cannot use test data statistics during training**.

````python
# Calculate feature standard deviation & mean
aq_sol_feature_std = np.std(aq_sol_feature_values, axis=0)
aq_sol_feature_mean = np.mean(aq_sol_feature_values, axis=0)

# Standardise the features
aq_sol_std_features = (aq_sol_feature_values - aq_sol_feature_mean) / aq_sol_feature_std

# Initialise a the weight vector with random values from a normal distribution. 
# The size of the vector is equal to the number of features
# We have to scale the initial weights by 0.1 so that the dot product of the weights and the new
# standardised features is not too small or too large
weight_vector = np.random.normal(scale=0.1, size=aq_sol_feature_dim)

# Initialise the bias term as 0.0
bias_term: float = 0.0

# Initialise number of epochs (how many times to iterate over all dataset batches
num_epochs: int = 3

loss_progress: List = []
learning_rate = 1e-2
batch_size: int = 32
num_data_points: int = len(aq_sol_label_values)

# Reshape `aq_sol_std_features` and `aq_sol_label_values` matrices/DataFrames to have 
# dimensions of `(-1, batch_size, feature_dim)`. The `-1` allows Numpy to automatically
# calculate the appropriate size for this dimension
aq_sol_batched_features = aq_sol_std_features[:num_batch_data].reshape(
    (-1, batch_size, aq_sol_feature_dim)
)
aq_sol_batched_labels = aq_sol_label_values[:num_batch_data].reshape(
    (-1, batch_size)
)

indices = np.arange(num_batch_data // batch_size)

# Iterate over all dataset batches 3 times
for epoch in range(num_epochs):
    
    # Iterate over batches randomly for stochastic gradient descent
    np.random.shuffle(indices)
    for i in indices:
        # Calculate gradients w.r.t weights and bias term
        grad = loss_grad(
            weight_vector, 
            bias_term, 
            aq_sol_batched_features[i], 
            aq_sol_batched_labels[i]
        )
    
        # Update weight vector by subtracting the product of the learning rate and the gradient
        # w.r.t the weights. This moves the weights in the direction of the negative gradient,
        # reducing the loss
        weight_vector -= learning_rate  * grad[0]
    
        # Similarly, update bias term by subtracting the product of the learning rate and the 
        # gradient w.r.t the bias
        bias_term -= learning_rate  * grad[1]
    
        # Calculate new MSE loss on each iteration to track loss progress every 50 steps
        if i % 50 == 0:
            loss_progress.append(
                linear_regression_utils.mean_squared_error_loss_wrapper(
                    weight_vector, 
                    bias_term,
                    aq_sol_std_features,
                    aq_sol_label_values
                )
            )
            
# Scale x-axis coordinates by 50 to include every step as we only record loss every 50 steps. 
# Keep y-axis coordinates the same
plt.plot(np.arange(len(loss_progress)) * 50, loss_progress)
plt.xlabel("Step")
plt.yscale("log")
plt.ylabel("Loss")
plt.title("Standardised Batched AqSolDB Training Curve")
plt.show()
````

## 4.6) Analysing Model Performance

### 4.6.1) Parity Plot

Analysing model performance is a **large topic**, however the first thing we typically examine in supervised learning is a **parity plot**, which plots the **model's predictions vs the label predictions**.

A parity plot is useful because it works **no matter what the feature dimensions are**. 

In a **perfect fit** all points would **lie on the line $y = \hat{y}$.

````python
linear_regression_predicted_labels = linear_regression_utils.linear_regression(
    aq_sol_std_features,
    weight_vector,
    bias_term
)

# Plot 45-deg reference line from (-100, -100) to (100, 100)
plt.plot([-100, 100], [-100, 100])

plt.scatter(aq_sol_label_values, linear_regression_predicted_labels, s=4, alpha=0.7)
plt.xlabel(r"Measured Solubility $y$")
plt.ylabel(r"Predicted Solubility $\hat{y}$")
plt.xlim(-13.5, 2)
plt.ylim(-13.5, 2)
plt.show()
````

### 4.6.2) Correlation Coefficient

As well as a parity plot, final model assessment can be done via **evaluation of the model's performance using the loss function**, as we have done earlier.

In regression, a **correlation coefficient** is typically reported **in addition to loss**.
* A correlation coefficient is a statistical measure that **quantifies the strength and direction** of the **linear relationship between two variables**.

Values for the correlation coefficient $r$ include:
* $r = 1$ which indicates that **as $x$ increases, $y$ increases perfectly in a linear fashion** (a **perfect positive linear relationship**)
* $r = -1$ which indicates that **as $x$ increases, $y$ decreases perfectly in a linear fashion** (a **perfect negative linear relationship**)
* $r = 0$ which indicates **no linear relationship between x and y**

We will use `np.correcoef()` to obtain the Pearson correlation coefficient between two variables. This returns a **correlation matrix**, which is a **2x2 matrix/array**, showing the **correlations between each pair of the input arrays**:

$$
\begin{bmatrix}
1 & r \\
r & 1
\end{bmatrix}
$$

where:
- $r$ is the **Pearson correlation coefficient between the two variables**.
- $1$ on the diagonal indicates the **correlation of each variable with itself**.

So for the below example, `np.corrcoef(aq_sol_label_values, linear_regression_predicted_labels)` would give:

```
array([[1.        , 0.69923293],
       [0.69923293, 1.        ]])
```

A **correlation coefficient of 0.70 is good, but not great**

````python
# Slice correlation between predicted labels and actual labels from correlation matrix
np.corrcoef(aq_sol_label_values, linear_regression_predicted_labels)[0, 1]
````
